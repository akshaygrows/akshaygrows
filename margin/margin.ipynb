```python
df_mm_lane = pd.read_csv('mm-cost.csv',header='infer')
df_mm_lane.drop(columns=['Orders (computable)'],inplace = True)
df_mm_lane['Cost'] = df_mm_lane['Cost'].str.replace(',', '').astype(int)
df_mm_lane.head()
```

### getting mm fm cost by lane

```python
cost_order_by_lane.head()
```

```python
cost_order_by_lane['self_orders'] = cost_order_by_lane['warehouse_city'] == cost_order_by_lane['city']
cost_order_by_lane['intercity_orders'] = cost_order_by_lane['warehouse_city'] != cost_order_by_lane['city']
cost_order_by_lane
```

```python
self_orders_sum = cost_order_by_lane[cost_order_by_lane['self_orders']].groupby('warehouse_city')['awb'].sum().reset_index()
intercity_orders_sum = cost_order_by_lane[cost_order_by_lane['intercity_orders']].groupby('warehouse_city')['awb'].sum().reset_index()

# Merge the results
mm_orders = pd.merge(self_orders_sum, intercity_orders_sum, on='warehouse_city', how='outer', suffixes=('_self_orders', '_intercity_orders')).fillna(0)

# Rename columns for clarity
mm_orders.columns = ['warehouse_city', 'awb_self_orders', 'awb_intercity_orders']

print(mm_orders)
```

```python
df_mm_lane_self = df_mm_lane[df_mm_lane['Lane Type']=='Hyperlocal']
df_mm_self = df_mm_lane_self.groupby('Warehouse City').aggregate({'Cost':'sum'}).reset_index()
df_mm_self.columns = ['warehouse_city','mm_cost']
df_mm_self
```

```python
df_mm_self_cpo = mm_orders.merge(df_mm_self,on='warehouse_city',how='left')
df_mm_self_cpo['self_mm_fm_cpo']=2*df_mm_self_cpo['mm_cost']/(2*df_mm_self_cpo['awb_self_orders'] + df_mm_self_cpo['awb_intercity_orders'])
df_mm_self_cpo['intercity_fm_cpo']=df_mm_self_cpo['mm_cost']/(2*df_mm_self_cpo['awb_self_orders'] + df_mm_self_cpo['awb_intercity_orders'])
df_mm_self_cpo
```

```python
df_mm_lane.columns = ['month','warehouse_city','shipping_city','vendor_name','cost','lane_type']
cost_order_by_lane.columns = ['warehouse_city','shipping_city','awb','self_orders','intercity_orders']
intercity_orders_by_lane = cost_order_by_lane[cost_order_by_lane['intercity_orders'] == True]
df_mm_intercity_cpo = intercity_orders_by_lane.merge(df_mm_lane,on=['warehouse_city','shipping_city'],how='left')
df_mm_intercity_cpo['intercity_mm_cpo'] = df_mm_intercity_cpo['cost']/df_mm_intercity_cpo['awb']
df_mm_intercity_cpo
```

```python
hyperlocal_df = rider_cost_df[rider_cost_df['warehouse_city']==rider_cost_df['city']]
hyperlocal_df = hyperlocal_df.merge(df_mm_self_cpo[['warehouse_city','self_mm_fm_cpo']], on ='warehouse_city', how='left')
hyperlocal_df['intercity_fm_cpo'] = 0
hyperlocal_df['intercity_mm_cpo'] = 0

intercity_df = rider_cost_df[rider_cost_df['warehouse_city']!=rider_cost_df['city']]
intercity_df = intercity_df.merge(df_mm_self_cpo[['warehouse_city','intercity_fm_cpo']], on ='warehouse_city', how='left')
intercity_df = intercity_df.merge(df_mm_intercity_cpo[['warehouse_city','shipping_city','intercity_mm_cpo']], left_on = ['warehouse_city','city'], right_on=['warehouse_city','shipping_city'],how='left')
intercity_df['self_mm_fm_cpo'] = 0

final_cost_df = pd.concat([hyperlocal_df, intercity_df]).drop(columns=['shipping_city']).reset_index()
final_cost_df['total_mm_cost'] = final_cost_df['self_mm_fm_cpo']+final_cost_df['intercity_mm_cpo']+final_cost_df['intercity_fm_cpo']
final_cost_df.drop(columns=['self_mm_fm_cpo','intercity_mm_cpo','intercity_fm_cpo'],inplace=True)

print(final_cost_df)
```
